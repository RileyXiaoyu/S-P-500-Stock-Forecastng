# M2 Velocity

```python
from fredapi import Fred
fred = Fred(api_key= 'f6111cac3afbe3680d59f72643824ac2') #Enter your key
data = fred.get_series('M2V') #Enter table name corresponding to FRED site
data.tail(10)
```

```python
# inverse order 
data = data.iloc[::-1]
data
```

```python
import pandas as pd
SP500 = pd.read_csv("S&P 500 Data.csv")
SP500
```

```python
data = pd.DataFrame(data)
data['Date'] = data.index
data = data.reset_index()
data = data.drop(['index'],axis=1)
```

```python
data
```

```python
data['Date'] = data['Date'].astype('datetime64[ns]')
SP500['Date'] = SP500['Date'].astype('datetime64[ns]')
```

```python
merged_data = pd.merge(data, SP500, how='left', on = 'Date')
```

```python
merged_data.rename(columns = {0:'Velocity of M2 Money Stock'}, inplace = True)
```

merged_data['Velocity of M2 Money Stock'] = pd.to_numeric(merged_data['Velocity of M2 Money Stock'])
merged_data['Price'] = pd.to_numeric(merged_data['Price'])

```python
merged_data
```

```python
import numpy as np
pearsons_coefficient = np.corrcoef(merged_data['Price'], merged_data['Velocity of M2 Money Stock'])
pearsons_coefficient
```

```python
merged_data.to_csv('Velocity of M2 Money Stock.csv',index=False)
```
